## 📊 Key Performance Indicators (KPIs)

<div style="background-color: var(--jp-content-background-color, #f8f9fa); color: var(--jp-content-font-color, #24292e); padding: 20px; border-radius: 10px; border-left: 5px solid #28a745; margin-top: 20px;">

### 💰 Revenue & Financial KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Total Revenue** | Sum of all transaction amounts | `Amount` |
| **Average Order Value (AOV)** | Total revenue ÷ Total orders | `Amount`, `Order ID` |
| **B2B vs B2C Revenue Split** | Revenue contribution by customer type | `Amount`, `B2B` |
| **Currency Distribution** | Revenue by different currencies | `Amount`, `currency` |

### 📦 Order & Fulfillment KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Order Volume** | Total number of orders | `Order ID` (count distinct) |
| **Order Status Breakdown** | Distribution of order statuses | `Status` |
| **Fulfillment Channel Performance** | Orders by fulfillment method | `Fulfilment`, `fulfilled-by` |
| **Shipping Service Levels** | Orders by shipping class | `ship-service-level` |
| **Courier Performance** | Orders by courier status | `Courier Status` |

### 📈 Product & Category KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Top Categories** | Revenue and quantity by category | `Category`, `Amount`, `Qty` |
| **Best Selling SKUs** | Top products by order volume | `SKU`, `Style`, `Qty` |
| **Size Distribution** | Order patterns by product size | `Size`, `Qty` |

### 🌍 Geographic KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Top Shipping Cities** | Orders by destination city | `ship-city` |
| **State-wise Performance** | Revenue by shipping state | `ship-state`, `Amount` |
| **International Orders** | Cross-border transaction analysis | `ship-country` |

### 📱 Sales Channel & Marketing KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Channel Performance** | Revenue by sales channel | `Sales Channel `, `Amount` |
| **Promotion Effectiveness** | Orders with promotions | `promotion-ids` |
| **B2B Penetration** | Percentage of B2B orders | `B2B`, `Order ID` |

### ⏱️ Temporal KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Daily/Monthly Trends** | Revenue patterns over time | `Date`, `Amount` |
| **Peak Season Analysis** | High-volume period identification | `Date`, `Order ID` |
