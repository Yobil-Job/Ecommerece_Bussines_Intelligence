## 📊 Key Performance Indicators (KPIs)

<div style="background-color: var(--jp-content-background-color, #f8f9fa); color: var(--jp-content-font-color, #24292e); padding: 20px; border-radius: 10px; border-left: 5px solid #28a745; margin-top: 20px;">

### 💰 Revenue & Financial KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Total Revenue** | Sum of all transaction amounts | `Amount` |
| **Average Order Value (AOV)** | Total revenue ÷ Total orders | `Amount`, `Order ID` |
| **B2B vs B2C Revenue Split** | Revenue contribution by customer type | `Amount`, `B2B` |
| **Currency Distribution** | Revenue by different currencies | `Amount`, `currency` |

### 📦 Order & Fulfillment KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Order Volume** | Total number of orders | `Order ID` (count distinct) |
| **Order Status Breakdown** | Distribution of order statuses | `Status` |
| **Fulfillment Channel Performance** | Orders by fulfillment method | `Fulfilment`, `fulfilled-by` |
| **Shipping Service Levels** | Orders by shipping class | `ship-service-level` |
| **Courier Performance** | Orders by courier status | `Courier Status` |

### 📈 Product & Category KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Top Categories** | Revenue and quantity by category | `Category`, `Amount`, `Qty` |
| **Best Selling SKUs** | Top products by order volume | `SKU`, `Style`, `Qty` |
| **Size Distribution** | Order patterns by product size | `Size`, `Qty` |

### 🌍 Geographic KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Top Shipping Cities** | Orders by destination city | `ship-city` |
| **State-wise Performance** | Revenue by shipping state | `ship-state`, `Amount` |
| **International Orders** | Cross-border transaction analysis | `ship-country` |

### 📱 Sales Channel & Marketing KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Channel Performance** | Revenue by sales channel | `Sales Channel `, `Amount` |
| **Promotion Effectiveness** | Orders with promotions | `promotion-ids` |
| **B2B Penetration** | Percentage of B2B orders | `B2B`, `Order ID` |

### ⏱️ Temporal KPIs
| KPI | Description | Data Source |
|:----|:------------|:------------|
| **Daily/Monthly Trends** | Revenue patterns over time | `Date`, `Amount` |
| **Peak Season Analysis** | High-volume period identification | `Date`, `Order ID` |

# PHASE 2 — DATA CLEANING (ADVANCED & STRUCTURED)

```python
Amazon_Sale_Report.drop(columns=['Unnamed: 22','index'],inplace=True)
```

```python
Amazon_Sale_Report.head(5)
```

```python
Amazon_Sale_Report.columns
```

```python
Amazon_Sale_Report.info()
```

```python
Amazon_Sale_Report['Date'] = pd.to_datetime(
    Amazon_Sale_Report['Date']
)
```

```python
Amazon_Sale_Report['Amount'] = pd.to_numeric(
    Amazon_Sale_Report['Amount']
)
```

```python
Amazon_Sale_Report['Qty'] = pd.to_numeric(
    Amazon_Sale_Report['Qty']
)
```

```python
Amazon_Sale_Report[['Order ID','Date','SKU','Qty','Amount']].isnull().sum()
```

```python
Amazon_Sale_Report = Amazon_Sale_Report.dropna(subset=['Amount'])
```

```python
Amazon_Sale_Report[['Order ID','Date','SKU','Qty','Amount']].isnull().sum()
```

```python
Amazon_Sale_Report.duplicated().sum()
```

```python
Amazon_Sale_Report.drop_duplicates(inplace=True)
```

```python
Amazon_Sale_Report.duplicated().sum()
```

```python
Amazon_Sale_Report['Year'] = Amazon_Sale_Report['Date'].dt.year
Amazon_Sale_Report['Month'] = Amazon_Sale_Report['Date'].dt.month
Amazon_Sale_Report['Month_Name'] = Amazon_Sale_Report['Date'].dt.month_name()
Amazon_Sale_Report['Quarter'] = Amazon_Sale_Report['Date'].dt.quarter
Amazon_Sale_Report['Day'] = Amazon_Sale_Report['Date'].dt.day
```

```python
Amazon_Sale_Report.head(5)
```

```python
Amazon_Sale_Report['Revenue'] = Amazon_Sale_Report['Amount']
```

```python
Expense_IIGF = Expense_IIGF.drop(0)
```

```python
Expense_IIGF = Expense_IIGF[
    Expense_IIGF['Income_Particular'] != 'Total'
]
```

```python
Expense_IIGF['Income_Amount'] = pd.to_numeric(
    Expense_IIGF['Income_Amount'],
    errors='coerce'
)

Expense_IIGF['Expense_Amount'] = pd.to_numeric(
    Expense_IIGF['Expense_Amount'],
    errors='coerce'
)
```

```python
Expense_IIGF
```

```python
total_income = Expense_IIGF['Income_Amount'].sum()
print(f'Total income: {total_income}')
```

```python
total_expense = Expense_IIGF['Expense_Amount'].sum()
print(f'Total Expence: {total_expense}')
```

```python
net_profit = total_income - total_expense
print(f'Net_Profit: {net_profit}')
```

```python
profit = total_revenue - total_expense
profit_margin = profit / total_revenue
print(f'Profit: {profit}')
print(f'Profit Margin: {profit_margin}')
 
```

# PHASE 5 — PROFESSIONAL VISUALIZATION

```python
import matplotlib.pyplot as plt

monthly_rev = (
    Amazon_Sale_Report
    .groupby(['Year','Month'])['Revenue']
    .sum()
    .reset_index()
)

monthly_rev['Date'] = pd.to_datetime(
    monthly_rev['Year'].astype(str) + '-' + monthly_rev['Month'].astype(str)
)

plt.figure()
plt.plot(monthly_rev['Date'], monthly_rev['Revenue'])
plt.title("Monthly Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
monthly_rev['Growth_Rate (%)'] = monthly_rev['Revenue'].pct_change() * 100

plt.figure()
plt.plot(monthly_rev['Date'], monthly_rev['Growth_Rate (%)'])
plt.title("Monthly Growth Rate (%)")
plt.xlabel("Date")
plt.ylabel("Growth Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
